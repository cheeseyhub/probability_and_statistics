```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt;
```

```python
dataframe = pd.read_csv("data/nhanes_2015_2016.csv")
```

```python
gender = dataframe["RIAGENDR"]
print(gender.value_counts())
```

```python
gender.value_counts().plot(kind="bar")
plt.xticks([0,1],["male","female"])
plt.xlabel("Gender")
plt.ylabel("Frequency")
plt.show()
```

```python
import random
sizes = np.array([random.randrange(40,55) for i in range(100)])
print(sizes.mean())
```

```python
from collections import Counter
counter = Counter();
for size in sizes:
    counter[size] +=1;
print(counter.most_common())
```

```python
from scipy import stats
print(stats.mode(gender))
```

```python
weight = dataframe["BMXWT"]
```

```python
weight.hist(bins=20);
```

```python
plt.hist(weight,bins=20)
plt.xlabel("weights")
plt.ylabel("Frequency")
plt.title("Histograms of weights in the data set")
plt.show()
```

```python
import seaborn as sns
sns.set_palette(None,color_codes=True)
sns.set_style("white")
```

```python
sns.histplot(weight,kde=True,bins=20)
plt.ylabel("Frequency")
plt.xlabel("Weights")
plt.title("Histogram of weights in NHANES Dataset using seaborn")
sns.despine(offset=5,trim=True)
plt.show()
```
