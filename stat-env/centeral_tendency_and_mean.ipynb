```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np  
```

```python
import random
nums = np.array([random.randrange(0,100) for i in range(200)])
```

```python

print(np.mean(nums));
print(np.median(nums));
```

```python
data = pd.read_csv("data/nhanes_2015_2016.csv");
```

```python
weight = data["BMXWT"];
```

```python
print(weight.max())
print(weight.min())
print(weight.median())
```

```python
x= range(weight.size);
y= weight;

plt.scatter(x,y,marker="o",s=0.5);
plt.title("Weights of individuals ");
plt.xlabel("ID of individuals");
plt.ylabel("Weight (kg)");

x_c,y_c = ([0,6000],[weight.median(),weight.median()]);
plt.plot(x_c,y_c,color="red",linewidth=0.75)

x_c,y_c = ([0,6000],[weight.mean(),weight.mean()]);
plt.plot(x_c,y_c,color="green",linewidth=0.75)
```
